```python
        ###   FIZZBUZZ  ÖDEVİ   #####

fizzbuzz = []
for i in range(1, 100):
  if i % 3 == 0 and i % 5 == 0:
    i = "FizzBuzz"
    fizzbuzz.append(i)
  elif i % 3 == 0 and i % 5 != 0:
    i = "Fizz"
    fizzbuzz.append(i)
  elif i % 5 == 0 and i % 3 != 0:
    i = "Buzz"
    fizzbuzz.append(i)
  else:
    fizzbuzz.append(i)
print(fizzbuzz)
```
