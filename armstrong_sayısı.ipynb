```python
kontrol = 0
sayı = input("bir sayı girin : ")
if float(sayı) < 0:
  print("negatif bir sayı girdiniz")
elif not sayı.isdigit():
  print("float bir sayı girdiniz")
else:
  for i in sayı:
    kontrol += int(i)**len(sayı)
  if kontrol == int(sayı):
    print("girdiğiniz sayı armstrong sayısıdir")
  else:
    print("armstrong sayısı değildir")
```
