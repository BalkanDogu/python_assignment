```python
        ###   ASAL SAYI ÖDEVİ   #####

asal = int(input("bir sayı giriniz : "))
bölünen = 0
if asal > 1:
  for i in range(2, (asal+1)):
    if asal % i == 0:
      bölünen += 1
  if bölünen >= 3:
    print("girdiğiniz sayı asal bir sayı değildir")
  else:
    print("girdiğiniz sayı asal bir saydır")
```
